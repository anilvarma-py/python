# DECORATORS

```python
def decor(func):
    def wrap():
        print("============")
        func()
        print("============")
    return wrap
def print_text():
    print("hello world")
decorated=decor(print_text)
decorated()
```

```python
def decor(func):
    def wrap():
        print("===========")
        func()
        print("===========")
    return wrap
@decor
@decor
def print_text():
    print("heyyy vizag")
decorated = decor(print_text) 
decorated()
```

# RECURSION
RECURSION IS A VERY IMPORTANT PART OF FUNCTIONAL  PROGRAMMING .
THE FUNDAMENTAL PART OF RECURSION IS SELF-REFERENCE -FUNCTIONS CALLING THEMSELVES.IT IS USED TO SOLVE PROBLEMS THAT CAN BE BROKEN UP INTO EASIER SUB-PROBLEMS OF THE SAME TYPE.
THE CLASSIC EXAMPLE OF THE FUNCTION THAT IS IMPLEMENTED RECURSIVELY IS THE FACTORIAL FUNCTION,WHICH FINDS THE PRODUCT OF ALL THE POSITIVE INTIGERS BELOW A SPECIFIED NUMBER.

```python
def factorial(x):
    if x==1:
        return 1
    else :
        return x*factorial(x-1)
print(factorial(5))
```

```python
def factorial(x):
    return x * factorial(x-1)        #incorrect version of factorial
print(factorial(5))
```
